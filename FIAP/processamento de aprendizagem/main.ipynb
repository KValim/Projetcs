## feature engineering

```python
from datetime import datetime
import numpy as np

df = df_clean

# 1. Convertendo o tipo da coluna 'primeiraCompra' para datetime
df['primeiraCompra'] = pd.to_datetime(df['primeiraCompra'])

# 2. Engenharia de Recursos Temporais
# Data atual para cÃ¡lculo da diferenÃ§a
from datetime import datetime
data_atual = datetime.now()

# Criando uma nova coluna com o tempo desde a primeira compra em dias
df['tempoDesdePrimeiraCompra'] = (data_atual - df['primeiraCompra']).dt.days

df[['primeiraCompra', 'tempoDesdePrimeiraCompra']].head()
df['idadeEmpresa'] = datetime.today().year - df['anoFundacao']

# 7. CriaÃ§Ã£o de Indicadores Booleanos
df['temAtraso'] = df['maiorAtraso'] > 0


### Binning
# 1. Aplicar a transformaÃ§Ã£o logarÃ­tmica
df['valorSolicitado_log'] = np.log1p(df['valorSolicitado'])  # log(x + 1)

# 2. Definir os limites dos "bins" na escala logarÃ­tmica
bins_log = np.linspace(df['valorSolicitado_log'].min(), df['valorSolicitado_log'].max(), 5)

# 3. Criar "bins" para os dados originais com base nos limites logarÃ­tmicos
df['valorSolicitado_bin'] = pd.cut(df['valorSolicitado_log'], bins=bins_log, labels=['Muito Baixo', 'Baixo', 'MÃ©dio', 'Alto'])
###


### booleanos
# Mapeando os valores booleanos para 1 e 0
boolean_map = {True: 1, False: 0}

df['temAtraso'] = df['temAtraso'].map(boolean_map)
df['empresa_MeEppMei'] = df['empresa_MeEppMei'].map(boolean_map)
df['restricoes'] = df['restricoes'].map(boolean_map)

df[['temAtraso', 'empresa_MeEppMei', 'restricoes']].head()
###

df
```

## pca

```python
from typing import List, Tuple
from sklearn.decomposition import PCA

class PCAAnalysis:
    """
    Classe para realizar a AnÃ¡lise de Componentes Principais (PCA) e identificar as colunas mais importantes.
    
    Attributes:
        n_components (int): NÃºmero de componentes principais para a anÃ¡lise.
    """
    def __init__(self, n_components: int):
        self.n_components = n_components
        self.pca = None
        self.loadings = None

    def fit(self, data: pd.DataFrame):
        """
        Ajusta o modelo PCA aos dados.
        
        Args:
            data (pd.DataFrame): DataFrame contendo os dados para anÃ¡lise.
        """
        # Padronizar os dados
        scaler = StandardScaler()
        scaled_data = scaler.fit_transform(data)
        
        # Aplicar PCA
        self.pca = PCA(n_components=self.n_components)
        self.pca.fit(scaled_data)
        
        # Obter coeficientes dos vetores prÃ³prios (loadings)
        self.loadings = pd.DataFrame(
            self.pca.components_, columns=data.columns,
            index=[f'Principal Component {i+1}' for i in range(self.n_components)]
            )

    def get_important_features(self, threshold: float = 0.2):
        """
        ObtÃ©m as caracterÃ­sticas mais importantes para cada componente principal com base em um limiar.
        
        Args:
            threshold (float): Valor absoluto do coeficiente acima do qual uma caracterÃ­stica Ã© considerada importante.
            
        Returns:
            List[Tuple[str, List[str]]]: Lista contendo tuplas. Cada tupla contÃ©m o nome do componente principal e a lista
                                            de colunas importantes para aquele componente.
        """
        if self.loadings is None:
            return []
        
        important_features = []
        for component in self.loadings.index:
            important_cols = self.loadings.loc[component, self.loadings.loc[component].abs() >= threshold].index.tolist()
            important_features.append((component, important_cols))
        
        return important_features

# Criar uma instÃ¢ncia da classe com 3 componentes principais
pca_analysis = PCAAnalysis(n_components=3)

# Ajustar aos dados
pca_analysis.fit(df_normalized)

# Obter caracterÃ­sticas importantes para cada componente
important_features = pca_analysis.get_important_features(threshold=0.2)

important_features
```

## cotovelo

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Definindo o range de valores de K para testar
k_values = range(1, 11)

# Lista para armazenar as inÃ©rcias (somas das distÃ¢ncias quadradas atÃ© o centro do cluster mais prÃ³ximo)
inertia_values = []

# Loop para calcular a inÃ©rcia para diferentes valores de K
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df_normalized)
    inertia_values.append(kmeans.inertia_)

# Plotando o grÃ¡fico do mÃ©todo do cotovelo
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia_values, marker='o')
plt.title("MÃ©todo do Cotovelo para Determinar o NÃºmero Ideal de Clusters")
plt.xlabel("NÃºmero de Clusters (K)")
plt.ylabel("InÃ©rcia")
plt.grid(True)
plt.show()
```

## silhouette_score

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
k_values = range(2, 11)  # ComeÃ§ando de 2 para evitar erros

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    clusters = kmeans.fit_predict(df_normalized)
    silhouette_avg = silhouette_score(df_normalized, clusters)
    silhouette_scores.append(silhouette_avg)

# Plotar os Silhouette Scores
plt.figure(figsize=(10, 6))
plt.plot(k_values, silhouette_scores, marker='o')
plt.xlabel('NÃºmero de Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score para Diferentes Valores de K')
plt.grid(True)
plt.show()
```
