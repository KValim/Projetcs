# Exerc√≠cios - comparando LightGBM e XGBoost

```python
# bibliotecas
import numpy as np 
import pandas as pd 
import lightgbm as lgbm
import xgboost as xgb
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, StratifiedKFold, cross_val_score
from datetime import datetime
import warnings
warnings.filterwarnings("ignore")
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# dataset
df_census = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header=None)
df_census.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation',
                  'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 
                   'income']
df_census.head()
```

```python
df_census.info()
```

```python
# EDA
df_census = df_census.drop(['education'], axis=1)
df_census = pd.get_dummies(df_census)
df_census = df_census.drop('income_ <=50K', axis=1)
df_census.head()
```

```python
#features e labels
X = df_census.iloc[:,:-1]
y = df_census.iloc[:,-1]
```

### ToDo 1
Separe o conjunto de features com suas respectivas labels em conjunto de treino e teste na propor√ß√£o de 70%-30%. Use random_state=42 para reprodutibilidade.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### ToDo 2

Treine um modelo LightGBM e um modelo XGBoost nos dados acima gerados. Utilize a vers√£o padr√£o de ambos os modelos. Use a biblioteca datetime para computar o tempo de treinamento de cada modelo. Avalie ambos usando a acur√°cia. 

```python
import datetime
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# Treinando o modelo XGBoost
start = datetime.datetime.now()
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)
end = datetime.datetime.now()
print(f"Tempo de treinamento XGBoost: {end - start}")

y_pred_xgb = xgb_model.predict(X_test)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print(f"Acur√°cia do modelo XGBoost: {accuracy_xgb}")
```

```python
# Treinando o modelo LightGBM
start = datetime.datetime.now()
lgbm_model = LGBMClassifier()
lgbm_model.fit(X_train, y_train)
end = datetime.datetime.now()
print(f"Tempo de treinamento LightGBM: {end - start}")

y_pred_lgbm = lgbm_model.predict(X_test)
accuracy_lgbm = accuracy_score(y_test, y_pred_lgbm)
print(f"Acur√°cia do modelo LightGBM: {accuracy_lgbm}")
```

### ToDo 3

Modifique a fun√ß√£o grid_search, vista na aula de XGBoost, para otimizar os parametros de ambos os modelos. Agora, ela precisa permitir receber, al√©m do conjunto de treino e respectivas respostas, o conjunto de teste e respectivas respostas.  

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, KFold

def grid_search(params, model, X_train, y_train, X_test, y_test, random=False, kfold=5):
    
    if random:
        grid = RandomizedSearchCV(model, params, cv=kfold, n_jobs=-1, random_state=2, scoring='recall')
    else:
        grid = GridSearchCV(model, params, cv=kfold, n_jobs=-1, scoring='recall')
    
    grid.fit(X_train, y_train)

    best_params = grid.best_params_
    best_score = grid.best_score_

    print("Best params:", best_params)
    print("Best score: {:.5f}".format(best_score))

    y_pred = grid.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_pred)
    
    return best_params, best_score, test_accuracy
```

### ToDo 4
Otimize os hiperpar√¢metros que achar conveniente. Compare o tempo de treinamento e a acur√°cia de cada modelo

* Par√¢metros XGBoost -> [link](https://xgboost.readthedocs.io/en/stable/parameter.html)

* Par√¢metros LightGBM -> [link](https://lightgbm.readthedocs.io/en/latest/Parameters.html)

```python
# Par√¢metros para otimiza√ß√£o
lgbm_params = {
    'learning_rate': [0.01, 0.1, 0.3],
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7]
}

xgb_params = {
    'learning_rate': [0.01, 0.1, 0.3],
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7]
}


# Otimizando XGBoost
best_xgb_params, best_xgb_score, xgb_test_accuracy = grid_search(xgb_params, XGBClassifier(random_state=2, scale_pos_weight=10, verbosity=0, use_label_encoder=False), X_train, y_train, X_test, y_test)
print(f"Melhores par√¢metros XGBoost: {best_xgb_params}")
print(f"Melhor acur√°cia de valida√ß√£o cruzada XGBoost: {best_xgb_score}")
print(f"Acur√°cia no conjunto de teste XGBoost: {xgb_test_accuracy}")
```

```python

# Otimizando LightGBM
best_lgbm_params, best_lgbm_score, lgbm_test_accuracy = grid_search(lgbm_params, LGBMClassifier(random_state=2), X_train, y_train, X_test, y_test)
print(f"Melhores par√¢metros LightGBM: {best_lgbm_params}")
print(f"Melhor acur√°cia de valida√ß√£o cruzada LightGBM: {best_lgbm_score}")
print(f"Acur√°cia no conjunto de teste LightGBM: {lgbm_test_accuracy}")
```

#### ToDo 5

Apenas para LightGBM. Use os parametros encontrados pelo boosting e treine 4 versoes diferentes usando os diferentes tipos de boosting (gbdt, goss, dart e rf).

* Obs: quando boosting_type=rf, use dois parametros adicionais

    -> bagging_freq=1
    
    -> bagging_fraction=0.8

```python
boosting_types = ['gbdt', 'goss', 'dart', 'rf']
lgbm_models = []

for boosting_type in boosting_types:
    params = best_lgbm_params.copy()
    params['boosting_type'] = boosting_type
    
    if boosting_type == 'rf':
        params['bagging_freq'] = 1
        params['bagging_fraction'] = 0.8
    
    model = LGBMClassifier(**params)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    
    lgbm_models.append((model, accuracy))
    print(f"Acur√°cia do LightGBM com boosting_type={boosting_type}: {accuracy}")

# Imprimir a acur√°cia de todos os modelos treinados
for i, (model, accuracy) in enumerate(lgbm_models):
    print(f"Acur√°cia do LightGBM com boosting_type={boosting_types[i]}: {accuracy}")
```
