```python
import flatten_json

# Get the "sales" collection
collection = db['sales']

# Use find method to retrieve all the documents
cursor = collection.find({})

# Create a list to hold the flat dictionaries
flattened_docs = []

# Iterate through the cursor and convert each document to a flat dictionary
for doc in cursor:
    flatten_doc = flatten_json.flatten(doc)
    flattened_docs.append(flatten_doc)

# Create a DataFrame from the list of flat dictionaries
df = pd.DataFrame(flattened_docs)

df
```
