```python
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import pandas as pd   
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
```

```python
cadastro = pd.read_excel('Prototipo_cadastro.xlsx')
cadastro
```

```python
compras = pd.read_excel('Prototipo_compras.xlsx')
compras.head()
```

```python
cadastro['IDADE'] = 2022 - cadastro['DATA_NASCIMENTO'].dt.year
cadastro['IDADE_PET'] = 2022 - cadastro['DATA_NASCIMENTO_PET'].dt.year
cadastro
```

```python
#Criando a chave primária 
cadastro['chave'] = cadastro['ID'].astype(str)+'-'+cadastro['NOME_PET']
compras['chave'] = compras['ID_PRODUTO'].astype(str)+'-'+compras['DATA'].astype(str)+'-'+compras['USUARIO'].astype(str)
```

```python

 
le = LabelEncoder() 
sexo_l = le.fit_transform(cadastro['SEXO'])
estado_l = le.fit_transform(cadastro['ESTADO'])
raca_l = le.fit_transform(cadastro['RAÇA_PET'])
sexo_pet_l = le.fit_transform(cadastro['SEXO_PET'])
 
cadastro.drop("SEXO", axis=1, inplace=True)
cadastro.drop("ESTADO", axis=1, inplace=True)
cadastro.drop("RAÇA_PET", axis=1, inplace=True)
cadastro.drop("SEXO_PET", axis=1, inplace=True)

cadastro["SEXO"] = sexo_l
cadastro["ESTADO"] = estado_l
cadastro["RACA_PET"] = raca_l
cadastro["SEXO_PET"] = sexo_pet_l
  
cadastro
```

```python
features = cadastro[["SEXO", "ESTADO", "RACA_PET", "SEXO_PET", "IDADE", "IDADE_PET"]]   
features
```

```python
pred_y = kmeans.fit_predict(features)
cadastro['CATEGORIA'] = pred_y
cadastro
```

```python
compras_prod = compras[["USUARIO", "PET","ID_PRODUTO"]]
compras_prod = compras_prod.drop_duplicates()
compras_prod = pd.get_dummies(compras_prod, columns = ['ID_PRODUTO'])
compras_prod['chave'] = compras_prod['USUARIO'].astype(str)+'-'+compras_prod['PET']

compras_prod=compras_prod.merge(cadastro[['chave','CATEGORIA']],how="left",on="chave")
compras_prod
```

```python
compras_prod_agg = compras_prod.groupby(["CATEGORIA"]).sum().reset_index()
compras_prod_agg = compras_prod_agg.drop(columns = ["USUARIO"])
compras_prod_agg2 = compras_prod_agg.drop(columns = ["CATEGORIA"])
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
scaled_x = min_max_scaler.fit_transform(compras_prod_agg2)
scaled_x

num_col = scaled_x.shape[1]

for i in range(0, num_col, 1):
  nome_col = 'ID_PRODUTO_'+str(i+1)
  nome_col2 = 'NOTA_CATEGORIA_PRODUTO_'+str(i+1)
  compras_prod_agg = compras_prod_agg.drop(columns = [nome_col])
  compras_prod_agg[nome_col2] = scaled_x[:,i]*10

df_USUARIOS = cadastro[['chave','ID','NOME_PET','CATEGORIA']]
df_USUARIOS=df_USUARIOS.merge(compras_prod_agg,how="left",on="CATEGORIA")
df_USUARIOS
```

```python
compras["MES"] = compras['DATA'].dt.month
compras["ANO"] = compras['DATA'].dt.year
compras['chave'] = compras['USUARIO'].astype(str)+'-'+compras['PET']

compras_atual = compras.query("ANO == 2022 and MES == 11")
compras_atual_prod = compras_atual.groupby(["ID_PRODUTO"]).chave.count().reset_index()
compras_atual_prod_scal = compras_atual_prod.drop(columns=["ID_PRODUTO"])
compras_atual_prod
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
scaled_x = min_max_scaler.fit_transform(compras_atual_prod_scal)
NOTA_Popularidade = scaled_x*10
NOTA_Popularidade
```

```python
num_col = len(NOTA_Popularidade)
num_col
tam= len(df_USUARIOS)

for i in range(1, num_col+1, 1):
  nome_col = 'NOTA_Pop_PRODUTO_'+str(i)
  nome_col2 = 'NOTA_CATEGORIA_PRODUTO_'+str(i)
  nome_col3 = 'NOTA_FINAL_'+str(i)
  df_USUARIOS[nome_col] = np.repeat(NOTA_Popularidade[i-1],tam)
  df_USUARIOS[nome_col3] = (df_USUARIOS[nome_col] + df_USUARIOS[nome_col2]) / 2

df_USUARIOS

df_USUARIOS['RECOMEND_1'] = 0
df_USUARIOS['RECOMEND_2'] = 0
df_USUARIOS['RECOMEND_3'] = 0
```

```python
num_col
```

```python
# REALIZANDO O LAÇO DE REPETIÇÃO PARA ENTENDER QUAL VAI SER A RECOMENDAÇÃO
for index, row in df_USUARIOS.iterrows():

  # CRIANDO VARIAVEIS PARA AS NOTAS
 df_nota = pd.DataFrame()
 df_nota['PRODUTO'] = ''
 df_nota['NOTA'] = ''

 for i in range(1, num_col+1, 1):
  nome_col = 'NOTA_FINAL_'+str(i)
 # CRIANDO UMA LISTA PARA ADICIONARMOS AO DF_NOTA
  lista_appendada = []

  # ADICIONANDO OS VALORES A UMA VARIAVEL  
  nota_x = df_USUARIOS[nome_col][index]
  produto_x = i

  # ADICIONANDO OS VALORES A UMA LISTA
  lista_appendada = [produto_x, nota_x]

  # ADICIONANDO AO DF
  a_series = pd.Series(lista_appendada, index = df_nota.columns)
  df_nota = df_nota.append(a_series, ignore_index = True)

 df_nota2 = df_nota.sort_values(by = 'NOTA', ascending=False) 
 df_nota2 = df_nota2.reset_index()
  # ADICIONAR O VALOR NA COLUNA DESEJADA
 df_USUARIOS['RECOMEND_1'][index] = df_nota2['PRODUTO'][0]
 df_USUARIOS['RECOMEND_2'][index] = df_nota2['PRODUTO'][1]
 df_USUARIOS['RECOMEND_3'][index] = df_nota2['PRODUTO'][2]

df_USUARIOS

```
