```python
# y = mx + n
# y = mx^2 + nx + P

# m - n + p= 0 
# 0m - 0n + p= 1
# m + n + p= 2
# 4m + 2n + p= 1

import numpy as np
import matplotlib.pyplot as plt


x1, y1 = -1, 0 
x2, y2 = 0, 1
x3, y3 = 1, 2
x4, y4 = 2,1
x5, y5 = 4,5

x = [x1,x2,x3,x4,x5]
y = [y1,y2,y3,y4,y5]

A = np.array([[x1**2,x1,1],[x2**2,x2, 1], [x3**2,x3,1], [x4**2,x4,1], [x5**2,x5,1]])
y = np.array([[y1], [y2], [y3], [y4], [y5]])
(m,n,p) = np.dot(np.dot(np.linalg.pinv(np.dot(A.T,A)), A.T), y)

plt.figure(figsize=(10,5))

plt.scatter(x1,y1, color='green', label='Pares ordenados (x,y)')
plt.scatter(x2,y2, color='green')
plt.scatter(x3,y3, color='green')
plt.scatter(x4,y4, color='green')

v_x = np.linspace(min(x), max(x))

plt.plot(x,y, label='Gráfico que liga os pontos')
plt.plot(v_x, m*v_x**2 + n*v_x + p, label='Reta otimizada' )
plt.grid(True)
plt.box(False)
plt.legend()
plt.title("Interpolação sobre todos os pontos")
```

```python
# y = mx + n
# y = mx^2 + nx + P
# y = kx^3 + mx^2 + nx + P

# m - n + p= 0 
# 0m - 0n + p= 1
# m + n + p= 2
# 4m + 2n + p= 1

import numpy as np
import matplotlib.pyplot as plt


x1, y1 = -1, 0 
x2, y2 = 0, 1
x3, y3 = 1, 2
x4, y4 = 2,1
x5, y5 = 4,5
p = 1

x = [x1,x2,x3,x4,x5]
y = [y1,y2,y3,y4,y5]

A = np.array([[x1**3,x1**2,x1,p],[x2**3,x2**2,x2, p], [x3**3,x3**2,x3,p], [x4**3,x4**2,x4,p], [x5**3,x5**2,x5,p]])
y = np.array([[y1], [y2], [y3], [y4], [y5]])
(k, m,n,p) = np.dot(np.dot(np.linalg.pinv(np.dot(A.T,A)), A.T), y)

plt.figure(figsize=(10,5))

plt.scatter(x1,y1, color='green', label='Pares ordenados (x,y)')
plt.scatter(x2,y2, color='green')
plt.scatter(x3,y3, color='green')
plt.scatter(x4,y4, color='green')

v_x = np.linspace(min(x), max(x))

plt.plot(x,y, label='Gráfico que liga os pontos')
plt.plot(v_x, k*v_x**3 + m*v_x**2 + n*v_x + p, label='Reta otimizada' )
plt.grid(True)
plt.box(False)
plt.legend()
plt.title("Interpolação sobre todos os pontos")
```
