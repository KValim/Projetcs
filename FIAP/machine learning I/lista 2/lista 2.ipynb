# <h1>2. Importando o Dataset:

```python
coeficientes_logisticos_estimados = pd.read_csv(url, header=None)

# Nomeando colunas
columns = ['variavel', 'coeficientes_logistico_estimado_B', 'wald', 'significancia', 'R_correlacao_parcial', 'exp_B']
coeficientes_logisticos_estimados.columns = columns

# Alterando o tipo de dado
coeficientes_logisticos_estimados[columns[1:len(columns)]] = coeficientes_logisticos_estimados[columns[1:len(columns)]].apply(lambda x: x.str.replace(',', '.').astype(float), axis=1)


coeficientes_logisticos_estimados.at[2,'variavel']='Primeira faixa de tempo de emprego'
coeficientes_logisticos_estimados.at[3,'variavel']='Segunda faixa de tempo de emprego'
coeficientes_logisticos_estimados.at[18,'variavel']='Categoria cagada de CEP'
coeficientes_logisticos_estimados
```

```python
coeficientes_logisticos_estimados.info()
```

# <h1>3. Probabildiade na Regressão Logística:

```python
def checagem(id_df):
  id_df = pd.concat([coeficientes_logisticos_estimados['coeficientes_logistico_estimado_B'], id_df['flags']], axis=1)
  id_df['aux'] = id_df['coeficientes_logistico_estimado_B'] * id_df['flags']

  new_df = id_df.copy()
  return new_df
```

```python
def calculo_z(id_df):
  id_df = pd.concat([coeficientes_logisticos_estimados['coeficientes_logistico_estimado_B'], id_df['flags']], axis=1)
  id_df['aux'] = id_df['coeficientes_logistico_estimado_B'] * id_df['flags']

  z = id_df['aux'].sum()
  return z
```

```python
def probabilidade(z):
  prob = math.exp(z)/(1 + math.exp(z))
  return prob
```
