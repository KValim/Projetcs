# Getting started
Once installed, you just need to `import` the module. Since AutoViz does not displays plots automatically, you also need to run `%matplotlib inline` before using it on your data.

Let's get started and import autoviz and the HCC dataset, which we will use for this notebook:



```python
from autoviz.AutoViz_Class import AutoViz_Class
import warnings; warnings.filterwarnings("ignore")
%matplotlib inline
```

**Note:** Don't forget to load the HCC dataset. Here we will read the file directly from our GitHub repository. However, you can first download the file and then upload it to your working directory and use it as `AutoViz_Class().AutoViz('hcc.csv')`. See this post on different ways to load data into Google Colab https://towardsdatascience.com/7-ways-to-load-external-data-into-google-colab-7ba73e7d5fc7.

```python
AutoViz_Class().AutoViz('https://raw.githubusercontent.com/Data-Centric-AI-Community/awesome-data-centric-ai/master/medium/data-profiling-tools/data/hcc.csv')
```

# Additional Features


```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Data-Centric-AI-Community/awesome-data-centric-ai/master/medium/data-profiling-tools/data/hcc.csv')

AV = AutoViz_Class()
dft = AV.AutoViz(
    filename="",
    sep="",
    depVar="Outcome",
    dfte=df,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="png",
    max_rows_analyzed=200,
    max_cols_analyzed=30,
)
```
